## Création d'une collection "Movies" dans une base de données.

```python
# Importer la librairie pymongo
import pymongo
# Connexion à mon client local Mongo
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
# Connexion à ma base de données spécifiée
mydb = myclient["spinoff3"]
# Spécifier le futur nom de ma nouvelle collection
mycol = mydb["Movies"]
```

## Insertion des données dans ma nouvelle collection "Movies".

```python
# Importer MongoClient depuis la librairie pymongo
# Importer ma librairie csv
# Importer ma librairie json
from pymongo import MongoClient
import csv
import json
# Connexion à la base de données
client = MongoClient('localhost', 27017)
db = client["spinoff3"]
col = db["Movies"]
# Fonction pour ajouter un fichier csv à un dictionnaire avec certains paramètres
def csv_to_dict():
    reader = csv.DictReader(open('movies.csv', 'r', encoding="UTF-8"))
    result = {}
    for row in reader:
        key = row.pop('id int, movieId int, title text, genres text')
        result[key]= row
    return result
    
    # Insertion finale
    x=col.insert_one(csv_to_dict())
    print(x.inserted_id)
    
# Pour insérer une ligne

# et pour insérer plusieurs lignes, le code suivant dooit être éxecuté
from pymongo import MongoClient
import csv
# Lire le fichier csv comme étant une liste de listes
client = MongoClient('localhost', 27017)
db = client["spinoff3"]
col = db["Movies"]
with open('movies.csv', 'r', encoding="UTF-8") as read_obj:
    # Créer une variable csv_reader permettant de lire mon fichier csv
    csv_reader = csv.DictReader(read_obj)
    # Lire ma variable pour obtenir une liste de listes
    mylist = list(csv_reader)
    # print(list_of_rows)
    x = col.insert_many(mylist)
```

## Vérifier que ma nouvelle base de données a bien été créée.

```python
print("Liste des base de données après insertion des données dans ma nouvelle BDD MongoDB:")
print(client.list_database_names()) # Va chercher la liste des base de données présentes.
```
